```python
!pip install Basemap

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import numpy.ma as ma
from mpl_toolkits.basemap import Basemap
from skimage import measure
from matplotlib.colors import ListedColormap
from mpl_toolkits.axes_grid1 import make_axes_locatable
import matplotlib.colors as colors
from matplotlib.ticker import LogLocator

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
lats_s = []

for i in range (0, 360):
    number = -(89.875 - i*0.25)
    lats_s.append(number)

lats = lats_s
for i in reversed(range(360)):
    lats.append(-lats_s[i])


longs = []
for i in range (0, 1440):
    number =  0.125 + i*0.25
    longs.append(number)
```

```python
filenames = ['/kaggle/input/water-vapour/15-aug-wv.txt','/kaggle/input/water-vapour/25-aug-wv.txt','/kaggle/input/water-vapour/25-sept-wv.txt','/kaggle/input/water-vapour/25-oct-wv.txt','/kaggle/input/water-vapour/5-oct-wv.txt','/kaggle/input/water-vapour/15-oct-wv.txt','/kaggle/input/water-vapour/15-sept-wv.txt','/kaggle/input/water-vapour/5-aug-wv.txt','/kaggle/input/water-vapour/5-sept-wv.txt']
dates = ['Aug 15, 2015', 'Aug 25, 2015', 'Sept 25, 2015', 'Oct 25, 2015', 'Oct 5, 2015', 'Oct 15, 2015', 'Sept 15, 2015', 'Aug 5, 2015', 'Sept 5, 2015']
for j in range (0, 9):
    f = open(filenames[j], 'r')
    l = f.readlines()
    f.close()
    for i in range (0, 7):
        l.pop(0)
    data = []
    for line in l:
        vals = line.split('\t')
        vals.pop(0)
        data.append(vals)

    data = [[s.rstrip('\n') if idx == len(sublist) - 1 else s for idx, s in enumerate(sublist)] for sublist in data]
    min_idx = 1440
    l1 = []
    for i in data:
        i.pop(len(i)-1)
        if len(i) < min_idx:
            l1 = i
            min_idx = len(i)
    truncated_data = []
    for i in data:
        truncated_data.append(i[0:min_idx])
    truncated_data = [[float(element) for element in sublist] for sublist in truncated_data]
    sst = np.array(truncated_data, dtype = float)
    latitudes = np.array(lats, dtype=float)
    longitudes = np.array(longs, dtype=float)
    m = Basemap(projection='cyl', llcrnrlat=-89.875, 
            urcrnrlat=89.75, 
            llcrnrlon=0.125, 
            urcrnrlon=longitudes[min_idx-1],
           resolution='c')
    lon, lat = np.meshgrid(longitudes[0:min_idx], latitudes)
    xi, yi = m(lon, lat)
    cmaps = ['viridis', 'bone', 'hot', 'RdPu', 'Greens']
    for i in range (0, len(cmaps)):
        masked_sst = ma.masked_where(sst == -999, sst)
        cs = m.pcolor(xi,yi,masked_sst, vmin=np.min(masked_sst), vmax=np.max(masked_sst), cmap=cmaps[i])
        m.drawcoastlines()
        m.drawparallels(range(-90, 90, 30), labels=[1, 0, 0, 0], fontsize=10, linewidth = 0.25)
        m.drawmeridians(range(-180, 180, 45), labels=[0, 0, 0, 1], fontsize=10, linewidth = 0.25)
        cbar = m.colorbar(cs, location='bottom', pad="10%")
        cbar.set_label('kg/m\u00b2')

        plt.title('Columnar Water Vapour'+' '+'('+dates[j]+')')
        plt.savefig('{}-normal-{}.png'.format(dates[j],cmaps[i]), dpi=600)
        plt.clf()
```
